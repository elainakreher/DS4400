```python
import pandas as pd
import numpy as np

df = pd.read_csv("kc_house_data.csv")
df = df.drop(columns=["id", "date", "zipcode"])

y = df["price"]             
X = df.drop(columns=["price"])
```

## 5.1:

```python
# Computing mean, min, max, variance for each X feature
summary = X.agg(["mean", "min", "max", "var"]).T
summary
```

```python
lowest_mean_feature  = summary["mean"].idxmin()
highest_mean_feature = summary["mean"].idxmax()

lowest_mean_value  = summary.loc[lowest_mean_feature, "mean"]
highest_mean_value = summary.loc[highest_mean_feature, "mean"]

lowest_mean_feature, lowest_mean_value, highest_mean_feature, highest_mean_value
```

```python
lowest_var_feature  = summary["var"].idxmin()
highest_var_feature = summary["var"].idxmax()

lowest_var_value  = summary.loc[lowest_var_feature, "var"]
highest_var_value = summary.loc[highest_var_feature, "var"]

lowest_var_feature, lowest_var_value, highest_var_feature, highest_var_value
```

### Answer:

For the features considered, the feature with the lowest average value is longitude, with a mean of ~ −122.21, reflecting the negative longitude values in the dataset. The feature with the highest average value is sqft_lot, with a mean of ~ 15,106.97, indicating that lot sizes are large on average compared to other features. 

Waterfront has the lowest variance (~ 0.0075), which is expected since it is a binary feature indicating the presence of a waterfront. The feature with the highest variance is sqft_lot, with a variance of ~ 1.72 × 10⁹, showing substantial variability in lot sizes across the dataset.

```python
corr_with_price = X.corrwith(y).sort_values(ascending=False)
corr_with_price
```

### Answer:

The Pearson correlation coefficients show that all features are positively correlated with house price, meaning that increases in any of the considered features are associated with higher prices. 

The feature with the strongest positive correlation with price is sqft_living (correlation ≈ 0.70), indicating that larger living area is most strongly associated with higher house prices. 

Other features like grade, sqft_above, and bathrooms also show relatively strong positive correlations with price, while features like condition and longitude have weaker positive relationships. 

No features were found to have a negative correlation with the response variable.
