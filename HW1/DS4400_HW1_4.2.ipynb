## Step 1: Create 3 Random Matrices

```python
import numpy as np

# no fixed seed so different each run
rang = np.random.default_rng()
matrices = [rang.integers(-10, 11, size=(3,3)) for _ in range(3)]
```

## Step 2 & 3: Compute Inverse of Each Matrix & Check by Obtaining Identity Matrix

```python
inverses = []

for i, A in enumerate(matrices, start=1):
    print(f"\nMatrix {i}:\n{A}")
    try:
        A_inv = np.linalg.inv(A)
        inverses.append(A_inv)
        print("Inverse:\n", A_inv)

        print("Identity Matrix:\n", np.round(A @ A_inv, 6))
    except np.linalg.LinAlgError:
        inverses.append(None)
        print("Singular matrix: no inverse.")
```
